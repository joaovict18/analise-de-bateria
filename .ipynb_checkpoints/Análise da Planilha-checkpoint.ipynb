```python
import pandas as pd

planilha = pd.read_excel("Baterias de Voo_nova.xlsx")
display(planilha)
```

```python
mediaTempo = planilha["Tempo de voo (s)"].sum() / len(planilha["Tempo de voo (s)"])
print(f"Média de tempo de voo: {mediaTempo}s")

soma = 0
for tempo in planilha["Tempo de voo (s)"]:
    soma += pow((tempo - mediaTempo), 2)

variacao = soma / len(planilha["Tempo de voo (s)"])
print(f"Variação do tempo de voo: {variacao}")
```
